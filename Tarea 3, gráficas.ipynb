```python
"""
    #a. Haz un programa que genere un arreglo de 1000 numeros consecutivos
    #b. Aplica una funcion lineal a los 1000 numeros
    #c. Grafica el resultado con matplotlib
    #d. Ahora intenta otras cinco funciones, cuadratica, logaritmo, etc. 
    #e. Haz una grafica donde quepan los graficos de las cinco funciones.
"""


#a para la lista de 1000 numeros 
inicio = int(input("Ingrese un valor inicial: ")) 
final = int(input("Ingrese un valor final: ")) #Debe ser un valor que avance mil veces el inicial +1
paso = 1
print(list(range(inicio,final, paso)))

```

```python
#b y c.Para la regresión lineal y la gráfica
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression

processDf = pd.read_csv('Posi_consec.csv')
processDf.head()
fig = plt.figure


plt.plot(processDf['Pos'],processDf['num']) #Da los ejes escalados
plt.xlabel('Poscición')
plt.ylabel('Números consecutivos')
plt.grid() #plt.show() quita cuadricula

```

```python
#para la línea de la regresión 
nPos= processDf['Pos'].values.reshape(-1,1)
nCons = processDf['num'].values.reshape(-1,1)
linear_regressor = LinearRegression()
linear_regressor.fit(nPos, nCons)
nCons_pred = linear_regressor.predict(nPos)
nCons_pred

#y=mx+c
m = linear_regressor.coef_[0][0] #rescatar valores
c = linear_regressor.intercept_[0]
label = r'$NC = %0.4f*posición  %+0.4f$'%(m,c) #Genera la ecuación
print(label) #y=mx+c
fig = plt.figure

#plt.scatter(processDf['Pos'],processDf['num'])
plt.plot(processDf['Pos'],processDf['num'])
plt.plot(nPos, nCons_pred, color='red', label=label)
plt.xlabel('Poscición')
plt.ylabel('Números consecutivos')
plt.legend()
plt.grid()
```
